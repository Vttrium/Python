```python
import pyautogui as pa
import time
import tkinter as tk
from tkinter import messagebox
```

```python
def play_verificacao(listNumbers):
    for number in range(len(listNumbers)):
        time.sleep(2.7)
        pa.click(-875, 82)  # Mouse click at (-875, 82)
        time.sleep(2.7)
        pa.click(-101, 108)  # Mouse click at (-101, 108)
        time.sleep(2.7)
        pa.write(str(listNumbers[number]))  # Type: 3
        time.sleep(2.61)
        pa.hotkey('tab')  # Hotkey: tab
        time.sleep(2.8)
        pa.hotkey('enter')  # Hotkey: enter
        time.sleep(2.5)
        pa.hotkey('enter')  # Hotkey: enter
        time.sleep(2)
        pa.click(-302, 245)  # Mouse click at (-302, 245)
        time.sleep(2.7)
        pa.click(-327, 288)  # Mouse click at (-327, 288)
        time.sleep(2.8)
        pa.click(-702, 366)  # Mouse click at (-702, 366)
        time.sleep(2.7)
        pa.click(-735, 430)  # Mouse click at (-735, 430)
        time.sleep(2.00)
        pa.click(-813, 865)  # Mouse click at (-813, 865)
        time.sleep(2.7)
        pa.click(-871, 70)  # Mouse click at (-871, 70)
        
    root = tk.Tk()
    root.withdraw()  # Oculta a janela principal
    messagebox.showinfo("Finalização", "O processo foi concluído com sucesso!")
    root.destroy()  # Fecha a janela oculta
```

```python
# Inicializa a lista para armazenar os números
numbers = []

print("Digite os números para adicionar à lista. Digite 'N' para finalizar.")

while True:
    # Solicita entrada do usuário
    user_input = input("Número: ")
    
    # Verifica se o usuário deseja sair
    if user_input.upper() == 'N':
        break
    
    # Tenta converter a entrada em número
    try:
        number = int(user_input)  # Converte a entrada em inteiro
        if number in numbers:  # Verifica se o número já existe na lista
            print("Número duplicado! Não será adicionado.")
        else:
            numbers.append(number)
            print(f"Número {number} adicionado com sucesso.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número ou 'N' para sair.")

# Exibe a lista final
print("\nLista final:")
print(numbers)
play_verificacao(numbers)
```

```python
def preencher_campos(listCodigoQtd):
    """
    Função para preencher campos em uma interface com base na lista de códigos e quantidades.
    
    Parâmetros:
        listCodigoQtd (list): Lista de tuplas [(codigo, quantidade), ...]
    """
    inativados_duplicados = []  # Lista para armazenar valores duplicados/inativos

    for item in range(len(listCodigoQtd)):
        # Inserir o código no primeiro campo
        pa.click(-1373, 330)  # Clica no campo do primeiro input
        time.sleep(0.2)
        pa.hotkey('ctrl', 'a')  # Seleciona o texto atual
        pa.hotkey('ctrl', 'c')  # Copia o texto para o clipboard

        # Captura o valor do clipboard
        current_value = pe.paste().strip()  # Remove espaços ou quebras de linha
        
        # Verifica se o campo está preenchido
        if current_value != "":
            inativados_duplicados.append(current_value)
            pa.hotkey('ctrl', 'a')  # Seleciona todo o texto
            pa.hotkey('delete')  # Apaga o texto selecionado

        # Insere o novo código
        pa.write(str(listCodigoQtd[item][0]))  # Escreve o código
        time.sleep(1.5)
        pa.hotkey('tab')  # Avança para o próximo campo

        # Inserir a quantidade no segundo campo
        pa.hotkey('ctrl', 'a')  # Seleciona o texto atual no segundo campo
        pa.hotkey('ctrl', 'c')  # Copia o texto para o clipboard

        # Captura o valor do clipboard
        current_value = pe.paste().strip()  # Remove espaços ou quebras de linha

        # Verifica se o campo está preenchido
        if current_value != "":
            pa.hotkey('ctrl', 'a')  # Seleciona todo o texto
            pa.hotkey('delete')  # Apaga o texto selecionado

        # Insere a nova quantidade
        pa.write(str(listCodigoQtd[item][1]))  # Escreve a quantidade
        pa.hotkey('tab')  # Avança ou finaliza
        time.sleep(0.2)
        if item == len(listCodigoQtd) - 1:
            pa.hotkey('tab')
        pa.hotkey('enter')  # Confirma os dados
        time.sleep(1)   
        
        
        
    return inativados_duplicados  # Retorna os valores duplicados/inativos, caso existam
```

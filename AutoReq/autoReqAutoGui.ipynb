```python
import pyautogui as pa
import pyperclip as pe
import time
import pandas as pd
```

```python
excel_file = r'C:/Users/SINNCAPPLE/Documents/VSCodeProjects/PythonProjects/AutoReq/Exemplos/RequisiÃ§Ã£o de Insumos 13-12-2024 - Formatada - Copia.xlsx'

all_sheets = pd.read_excel(excel_file, sheet_name=None)
sheet_names = list(all_sheets.keys())

num_sheet = 1 # Numero da janela que vc estÃ¡ no momento, comeÃ§a na primeira janela como 1;

df = all_sheets[sheet_names[num_sheet - 1]]

Codigo = df.iloc[:, 3]
qtd = df.iloc[:, 4]
Codigo = Codigo.dropna(how='all', axis=0)
qtd = qtd.dropna(how='all', axis=0)
qtd = qtd.drop(index=0)

listCodigoQtd = []
count = 0
for codigos in (Codigo):
    listCodigoQtd.append([int(codigos)])
    count += 1
for num, qtds in enumerate(qtd):
    listCodigoQtd[num].append(round(qtds))
    
display(df)
print(listCodigoQtd)
```

```python
for item in range(len(listCodigoQtd)):
    # Inserir o cÃ³digo no primeiro campo
    pa.click(-1373, 330)  # Clica no campo do primeiro input
    time.sleep(0.2)
    pa.hotkey('ctrl', 'a')  # Seleciona o texto atual
    pa.hotkey('ctrl', 'c')  # Copia o texto para o clipboard
    
    # Captura o valor do clipboard
    current_value = pe.paste().strip()  # Remove espaÃ§os ou quebras de linha
    inativados_duplicados = []
    
    if current_value != "":  # Verifica se o campo estÃ¡ preenchido
        print(f"Campo preenchido com: {current_value}. Limpando...")
        inativados_duplicados.append(current_value)
        pa.hotkey('ctrl', 'a')  # Seleciona todo o texto
        pa.hotkey('delete')  # Apaga o texto selecionado
    
    # Insere o novo cÃ³digo
    pa.write(str(listCodigoQtd[item][0]))  # Escreve o cÃ³digo
    time.sleep(1.5)
    pa.hotkey('tab')  # AvanÃ§a para o prÃ³ximo campo
    
    # Inserir a quantidade no segundo campo
    pa.hotkey('ctrl', 'a')  # Seleciona o texto atual no segundo campo
    pa.hotkey('ctrl', 'c')  # Copia o texto para o clipboard
    
    # Captura o valor do clipboard
    current_value = pe.paste().strip()  # Remove espaÃ§os ou quebras de linha
    
    if current_value != "":  # Verifica se o campo estÃ¡ preenchido
        print(f"Campo preenchido com: {current_value}. Limpando...")
        pa.hotkey('ctrl', 'a')  # Seleciona todo o texto
        pa.hotkey('delete')  # Apaga o texto selecionado
    
    # Insere a nova quantidade
    pa.write(str(listCodigoQtd[item][1]))  # Escreve a quantidade
    pa.hotkey('tab')  # AvanÃ§a ou finaliza
    time.sleep(0.2)
    pa.hotkey('enter')  # Confirma os dados
    time.sleep(1)
```

Adicionar texto com edição dentro do texto

```python
documento = Document()

fat = 2000

texto = f"""Fala galera,

Aluakibar"""

paragrafo = documento.add_paragraph(texto)

#-------------------------#

paragrafo.add_run(f" R${fat}").bold = True

#-------------------------#

texto_final = "Tamo junto."

paragrafo = documento.add_paragraph(texto_final)

documento.save("Texto.docx")
```

### Controle de Margem e Seções

```python
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.enum.text import WD_TAB_ALIGNMENT
```

```python
documento = Document()

for secao in documento.sections:
    secao.top_margin = Cm(0.5)
    secao.bottom_margin = Cm(1)
    secao.left_margin = Cm(1)
    secao.right_margin = Cm(1)

fat = 2000

texto = f"""Fala galera,

Aluakibar fat = {fat}

kkk.
"""
paragrafo = documento.add_paragraph(texto)

#----------------------#
paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER # / RIGHT / CENTER / JUSTIFY
#----------------------#

documento.save("Texto.docx")
```

### Como Inserir Imagem:

```python
documento = Document()

for secao in documento.sections:
    secao.top_margin = Cm(0.5)
    secao.bottom_margin = Cm(1)
    secao.left_margin = Cm(1)
    secao.right_margin = Cm(1)

fat = 2000

texto = f"""Fala galera,

Aluakibar fat = {fat}

kkk.
"""
paragrafo = documento.add_paragraph(texto)
paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER

imagem = documento.add_picture('imagem.png', width= Cm(10), height= Cm(10))
imagem.alignment = WD_TAB_ALIGNMENT.CENTER

documento.save("Texto.docx")
```

### Como Inserir tabela:

```python
documento = Document()

for secao in documento.sections:
    secao.top_margin = Cm(0.5)
    secao.bottom_margin = Cm(1)
    secao.left_margin = Cm(1)
    secao.right_margin = Cm(1)

fat = 2000

texto = f"""Fala galera,

Aluakibar fat = {fat}

kkk.
"""
paragrafo = documento.add_paragraph(texto)
paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER

records = (
    (3, '101', 'Spam'),
    (7, '422', 'Eggs'),
    (4, '631', 'Spam, spam, eggs, and spam')
)

table = documento.add_table(rows=1, cols=3, style='Medium Grid 3 Accent 4')
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Qty'
hdr_cells[1].text = 'Id'
hdr_cells[2].text = 'Desc'
for qty, id, desc in records:
    row_cells = table.add_row().cells
    row_cells[0].text = str(qty)
    row_cells[1].text = id
    row_cells[2].text = desc


documento.save("Texto.docx")
```

### Adaptar um Documento:

Tem placeholders

```python
from datetime import datetime

nome = "Pedro"
item1 = "Serviço de treino"; item2 = "Apostila"; item3 = "Serviço de rastreio"

dicionario_valores = {
    "XXXX": nome,
    "YYYY": item1,
    "ZZZZ": item2,
    "WWWW": item3,
    "DD": str(datetime.now().day),
    "MM": str(datetime.now().month),
    "AAAA": str(datetime.now().year),
}
```

```python
contrato = Document('Contrato.docx')

for paragrafo in contrato.paragraphs:
    for item in dicionario_valores:
        if item in paragrafo.text:
            paragrafo.text = paragrafo.text.replace(item, dicionario_valores[item])

contrato.save("Contrato Com Dados.docx")
```

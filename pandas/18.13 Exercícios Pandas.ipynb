```python
import pandas as pd

funcionarios = pd.read_csv('CadastroFuncionarios.csv', sep = ';', decimal=',')
display(funcionarios)
servicos = pd.read_excel(r'BaseServiçosPrestados.xlsx')
display(servicos)
clientes = pd.read_csv(r'Contoso - Clientes.csv', sep=';', decimal=',')
display(clientes)
```

```python
qtde_funcionarios_area = funcionarios['Area'].value_counts()
```

```python
funcionarios = funcionarios[['Salario Base', 'Impostos', 'Beneficios']]
total = 0
total_s = 0
total_i = 0
total_b = 0
for salario in funcionarios['Salario Base']:
    total_s += float(salario)
for imposto in funcionarios['Impostos']:
    imposto = imposto.replace(',', '.')
    total_i += float(imposto)
for benef in funcionarios['Beneficios']:
    benef = benef.replace(',', '.')
    total_b += float(benef)
total = total_s + total_i + total_b
```

```python
funcionarios['Salario Total'] = funcionarios['Salario Base'] + funcionarios['Impostos'] + funcionarios['Beneficios'] + funcionarios['VT'] + funcionarios['VR']
print('Total de folha salarial é de R${:,.2f}'.format(sum(funcionarios['Salario Total'])))
```
